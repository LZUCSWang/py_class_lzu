```python
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Given text
text = """
Knowledge can be acquired from many sources. These include books, teachers and practical experience, and each has its own advantages. 
The knowledge we gain from books and formal education enables us to learn about things that we have no opportunity to experience in daily life. 
We can study all the places in the world and learn from people we will never meet in our lifetime, just by reading about them in books. 
We can also develop our analytical skills and learn how to view and interpret the world around us in different ways. 
Furthermore, we can learn from the past by reading books. In this way, we won't repeat the mistakes of others and can build on their achievements.

Practical experience, on the other hand, can give us more useful knowledge. 
It is said that one learns best by doing, and I believe that this is true, whether one is successful or not. 
In fact, I think making mistakes is the best way to learn. Moreover, if one wants to make new advances, it is necessary to act. 
Innovations do not come about through reading but through experimentation. 
Finally, one can apply the skills and insights gained through the study of books to practical experience, making an already meaningful experience more meaningful. 
However, unless it is applied to real experiences, book knowledge remains theoretical and, in the end, is useless. 
That is why I believe that knowledge gained from practical experience is more important than that acquired from books.
"""

# Convert text to lowercase and split into words
words = text.lower().split()

# List of stop words (commonly used words that are generally ignored in text analysis)
stop_words = set([
    "the", "and", "to", "of", "in", "a", "is", "that", "it", "for", "as", "with", "on", "be", 
    "this", "from", "by", "are", "can", "we", "an", "has", "have", "not", "but", "if", "or",
    "i", "you", "he", "she", "they", "us", "at", "one", "do", "does", "did", "will", "would", 
    "should", "could", "about", "all", "also", "through", "which", "more", "just", "other", "no", 
    "such", "only", "out", "up", "when", "who", "what", "how", "so", "than", "them", "their", 
    "was", "its", "our", "your", "his", "her", "him", "me", "my", "own", "don't", "won't", "can't",
    "because", "while", "between", "into", "over", "after", "before", "again", "here", "there", 
    "where", "why", "now", "then", "always", "never", "been", "being", "were", "am", "had", 
    "some", "these", "those", "very", "any", "every", "else", "each", "off", "most", "least", 
    "many", "much", "both", "either", "neither", "whether", "like", "unlike", "same", "different",
    "well", "bad", "good", "anyone", "someone", "everyone", "noone", "nothing", "something", 
    "anything", "thing", "things", "part", "another", "such", "rather", "quite", "too", "enough",
    "few", "fewer", "many", "several", "lot", "lots", "plenty", "various", "whatever", "whichever",
    "whenever", "wherever", "however", "although", "though", "since", "until", "during", "within",
    "without", "beyond", "towards", "against", "among", "across", "behind", "beside", "besides",
    "plus", "minus", "per", "upon", "concerning", "regarding", "toward", "onto", "aboard", "amid",
    "amidst", "amongst", "apropos", "atop", "barring", "circa", "despite", "except", "excepting",
    "excluding", "failing", "following", "including", "past", "pending", "pro", "sans", "save", "via"
])

# Remove stop words and non-alphabetic words
```

```python
# Remove stop words and non-alphabetic words from the list
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]

# Count the frequency of each word
word_freq = Counter(filtered_words)

# Get the most common 5 words
most_common_words = word_freq.most_common(10)

# Now let's create a word cloud with these words
# To give more importance to the most common words, we'll create a dictionary with words and their frequencies
word_freq_for_cloud = {word: freq for word, freq in most_common_words}

# Generate the word cloud
wordcloud = WordCloud(width = 800, height = 400, background_color ='white').generate_from_frequencies(word_freq_for_cloud)

# Display the word cloud using matplotlib
plt.figure(figsize = (8, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
